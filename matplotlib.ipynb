```python
import matplotlib.pyplot as plt
```

```python
# create first chart
```

```python
x= [1,2,3,4,5,6]
y = [ 10,20,27,40,45,60] 

plt.plot(x,y) # line chart
plt.show()   #display chart in jp notebook
```

```python
# customize chart
```

```python
plt.plot(x,y,color='yellow', marker='o', linestyle='dashed', linewidth=2, markersize=15)
plt.show()
```

```python
plt.plot(figures = (4,3))  # figure size

plt.plot(x,y,color='yellow', marker='o', linestyle='dashed', linewidth=2, markersize=15)


plt.title("jai ho")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
# adavnced line charts 
```

```python
x= [1,2,3,4,5,6]
y1 = [10,20,27,40,45,60] 
y2 = [12,14,25,37,45,50]
plt.plot(x,y1,label= 'Sales 2024')
plt.plot(x,y2,label= 'Sales 2025')

plt.title("YoY sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

```python
# used to show part to whole relationship/data
```

```python
Categories= [1,2,3,4,5,6]
Sales = [ 10,20,27,40,45,60] 

plt.pie(Sales, labels = Categories, autopct= '%1.1f%%' )
plt.title("pie chart example")
plt.show()
```

```python
Categories= [1,2,3,4,5,6]
Sales = [ 10,20,27,40,45,60] 

plt.pie(Sales, labels = Categories, autopct= '%1.1f%%', startangle = 90 )
plt.title("pie chart example")
plt.show()
```

```python
# sub parts
```

```python
# use to show multiple  charts in one figure
# data1 - bar chart
Categories= ['mon','tues','wed','thurs','fri','sat']
Sales = [ 10,20,27,40,45,60] 

# data2 - scatter plot
y1 = [10,20,27,40,45,60] 
y2 = [12,14,25,37,45,50]

plt.figure(figsize=(10,6))

# first plot - bar chart
plt.subplot(1,2,1)  #row,column ,position
plt.bar(Categories,Sales)

plt.title("daily sales")
plt.xlabel("week days")
plt.ylabel("sales")

# first plot - scatter chart
plt.subplot(1,2,2)  #row,column ,position
plt.scatter(y1,y2)

plt.title("user example")
plt.xlabel("user 1")
plt.ylabel("user 2")

plt.show()

```

```python
# matplotlib with pandas real data
```

```python
# create dataframe
import pandas as pd

data = {
    'Month' :['Jan','feb','mar','apr'],
    'sales' :[12000,11000,13000,25000]
}

df= pd.DataFrame(data)
df
```

```python
plt.bar(df['Month'],df['sales'])
plt.title("matplotlib with pandas")
plt.xlabel("Month")
plt.ylabel("sales")

plt.show()
```

```python
# save plots /charts
```

```python
plt.bar(df['Month'],df['sales'])
plt.title("matplotlib with pandas")
plt.xlabel("Month")
plt.ylabel("sales")

plt.show()
plt.savefig("monthly_sales_from_df.png")
```
